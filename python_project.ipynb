```python
# if you want to find correlation between all columns
# convert the other columns to data type category
```

```python
df_allnumcol=df
for col in df_allnumcol.columns:
    if (df_allnumcol[col].dtype=='object'):
        df_allnumcol[col]=df_allnumcol[col].astype('category')
        df_allnumcol[col]=df_allnumcol[col].cat.codes
df_allnumcol        
```

```python
# run correlation function to determine correlation between all columns
```

```python
df_corr=df_allnumcol.corr()
df_corr
```

```python
# it can also be plotted
```

```python
sns.heatmap(df_corr, annot=True)
plt.title('Heatmap of all columns')
plt.xlabel('All movie features')
plt.ylabel('All movie features')
```

```python
df_corr=df_allnumcol.corr().unstack()
df_corr
```
